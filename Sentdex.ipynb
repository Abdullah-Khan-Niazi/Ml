```python
def func(string):
    upper_case = 0
    lower_case = 0
    for char in string:
        if char.isupper():
            upper_case += 1
        elif char.islower():
            lower_case += 1
    return upper_case, lower_case
string = "Hello World"
upper, lower = func(string)
print("Upper Case Letters:", upper)
print("Lower Case Letters:", lower)
```

```python
def func(bytes):
    megabytes = bytes // (1024 * 1024)
    gigabytes = bytes // (1024 * 1024 * 1024)
    terabytes = bytes // (1024 * 1024 * 1024 * 1024)
    return megabytes, gigabytes, terabytes
bytes = 1234567890000000
mb, gb, tb = func(bytes)
print("Megabytes:", mb)
print("Gigabytes:", gb)
print("Terabytes:", tb)
```

```python
import numpy as np
import matplotlib.pyplot as plt

arr = np.random.randint(1,7,500)

count = [0,0,0,0,0,0]
for i in arr:
    if i == 1:
        count[i - 1] += 1
    elif i == 2:
        count[i - 1] += 1
    elif i == 3:
        count[i - 1] += 1
    elif i == 4:
        count[i - 1] += 1
    elif i == 5:
        count[i - 1] += 1
    elif i == 6:
        count[i - 1] += 1
for i in range(len(count)):
    print(f"Value {i+1} : {count[i]}")
print(f"Value {count.index(max(count)) + 1} Occurred Maximum Times With {max(count) * 100 / 500} %")

plt.hist(arr,bins = 15)
plt.xlabel('Dice Value')
plt.ylabel('Frequency')
plt.title('Dice Roll Frequency Histogram')
plt.xticks(np.arange(1, 7))
plt.show()
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.isna().sum()
```

```python
count = df[(df['Survived'] == 1) & (df['Age'] > 70) ]
print("Number of People Survived and Aged Above 70:",len(count))
```

```python
grouped_df = df.groupby('Sex')['Survived'].sum()
grouped_df
```

```python
print("Survival Rate For Males: ",grouped_df['male'] * 100 / (df['Sex'] == 'male').sum())
print("Survival Rate For Females: ",grouped_df['female'] * 100 / (df['Sex'] == 'female').sum())
```
